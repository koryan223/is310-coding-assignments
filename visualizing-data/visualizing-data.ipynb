```python
import pandas as pd
import altair as alt
from bs4 import BeautifulSoup
import requests
```

```python
data = pd.read_csv('cleaned_pudding_data.csv')
#read the data
data['year'] = data['year'].astype(int)
#check the data type of year column and convert it to integer
data_cleaned = data.dropna(subset=['gross (inflation-adjusted)'])

# Display cleaned data
print(data_cleaned.head())
print(data_cleaned.info())
```

```python
chart = alt.Chart(data_cleaned).mark_circle(size=60).encode(
    x=alt.X('year:Q', title='Year of Release', scale=alt.Scale(domain=(1930, 2010))),
    #set the x-axis to be the year of release so we can see the trend over time
    y=alt.Y('gross (inflation-adjusted):Q', title='Gross Revenue (Adjusted)', scale=alt.Scale(domain=(0, 1400))),
    tooltip=['title', 'year', 'gross (inflation-adjusted)']
).properties(
    title='Trend of Gross Revenue Over Time',
    width=600,
    height=400
).interactive()


chart.display()
```

```python
import numpy as np

def word_count(text):
    words = text.split()
    return len(words)

word_counts = []
for link in data_cleaned['link']:
    volume_url = link
    volume_response = requests.get(volume_url)
    if volume_response.status_code == 200:
        volume_soup = BeautifulSoup(volume_response.text, "html.parser")
        volume_text = volume_soup.get_text().replace("\n", "")
        count = word_count(volume_text)
        word_counts.append(count)
    else:
        word_counts.append(np.nan)
```

```python
data_cleaned['word_count'] = word_counts
data_cleaned
```

```python
data_cleaned=data_cleaned.dropna().reset_index(drop=True)
data_cleaned
```

```python
chart = alt.Chart(data_cleaned).mark_circle(size=60).encode(
    x=alt.X('year:Q', title='Year of Release', scale=alt.Scale(domain=(1930, 2010))),
    #set the x-axis to be the year of release so we can see the trend over time
    y=alt.Y('word_count:Q', title='word count'),
    tooltip=['title', 'year', 'gross (inflation-adjusted)']
).properties(
    title='Trend of Gross Revenue Over Time',
    width=600,
    height=400
).interactive()


chart.display()
```
